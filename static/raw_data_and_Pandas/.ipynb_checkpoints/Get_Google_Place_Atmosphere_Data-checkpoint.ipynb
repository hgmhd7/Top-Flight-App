```python
# Store place IDs into list for full API retrieval below
place_ID_list = []

for place_id in airports_place_id_df['google_place_id']:
    
    place_ID_list.append(place_id)
    
    
print(place_ID_list[:5])
```

# Get Googlel API data for airport ratings, and number of ratings

```python
# Set counter for API calls
counter = 0

# Set empty lists to store values
airport_rating_list = []
number_of_ratings_list = []


# loop through airport names to retrive the Google names, place IDs, and addresses
for place_ID in place_ID_list:
    
    try:
       # Set target airport place ID
        target_airport_place_ID = place_ID

        # Build the endpoint URL
        target_url = ('https://maps.googleapis.com/maps/api/place/details/json?&placeid={0}&fields=rating,user_ratings_total&key={1}').format(target_airport_place_ID, gkey) 


        # Run a request to endpoint and convert result to json
        airport_geo_data = requests.get(target_url).json()
        
        
        airport_rating = airport_geo_data['result']['rating']
        airport_number_ratings = airport_geo_data['result']['user_ratings_total']
       
        
        # Appent to lists
        airport_rating_list.append(airport_rating) 
        number_of_ratings_list.append(airport_number_ratings)
       
        
        counter += 1 
        
            
        print(f'\nRetriveing ratings data for airport number {counter}.\
               \nRatings data retrieval complete!!!')
        
        
    
    except (IndexError, KeyError):
        
        airport_rating_list.append('not found')
        number_of_ratings_list.append('not found')
       
    
        counter += 1
        print(f'\n ERROR retriveing data for airport number {counter}... Skipping...')
```

```python
# Check number of ratings list
number_of_ratings_list[:5]
```

```python
# Check list length
len(airport_rating_list)
```

```python
# Check list length
len(number_of_ratings_list)
```

```python
# Add ratings to DF
airports_place_id_df["airport_rating"] = airport_rating_list
```

```python
# Add number of ratings to DF
airports_place_id_df["number_ratings"] = number_of_ratings_list
```

# Get Google API data for airport websites

```python
# Set counter for API calls
counter = 0

# Set empty lists to store values
website_list = []


# loop through airport names to retrive the Google names, place IDs, and addresses
for place_ID in place_ID_list:
    
    try:
       # Set target airport place ID
        target_airport_place_ID = place_ID

        # Build the endpoint URL
        target_url = ('https://maps.googleapis.com/maps/api/place/details/json?&placeid={0}&fields=website&key={1}').format(target_airport_place_ID, gkey) 


        # Run a request to endpoint and convert result to json
        airport_geo_data = requests.get(target_url).json()
        
        
        airport_website = test_website = airport_geo_data['result']['website']
       
        
        # Appent to lists
        website_list.append(airport_website)
       
        
        counter += 1 
        
            
        print(f'\nRetriveing website data for airport number {counter}. \
             \nAirport website: {airport_website}.\
              \nWebsite data retrieval complete!!!')
        
        
    
    except (IndexError, KeyError):
        
        website_list.append('not found')
       
    
        counter += 1
        print(f'\n ERROR retriveing website data for airport number {counter}... Skipping...')
```
